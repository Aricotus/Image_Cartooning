```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
#loading the image
img=cv.imread("D:\Coding\img_project.jpg")
img=cv.cvtColor(img , cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#converting into gray scale,median blurring and creating edge mask
img_gray= cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gray_blur=cv.medianBlur(img_gray ,5)
plt.imshow(img_gray_blur ,cmap='gray')
plt.show()
edges=cv.adaptiveThreshold(img_gray_blur,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,9)
plt.imshow(edges,cmap='gray')
plt.show()
```

```python
#image quantization using K-Means Clustering 
Z=img.reshape((-1,3))
Z=np.float32(Z)
criteria=(cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER ,10,1.0)
K=20
ret,label1,center1=cv.kmeans(Z,K,None,criteria,10,cv.KMEANS_RANDOM_CENTERS)
center1=np.uint8(center1)
res1=center1[label1.flatten()]
img1=res1.reshape(img.shape)
plt.imshow(img1)
plt.show()
```

```python
cv.bilateralFilter(img1,9,300,300)
plt.imshow(img1)
plt.show()
```

```python
#merging edge mask and quantized image
img_cartoon=cv.bitwise_and(img1,img1,mask=edges)
plt.imshow(img_cartoon,cmap="gray")
plt.show()
```

```python
#saving the cartoonized image
img_cartoon=cv.cvtColor(img_cartoon , cv.COLOR_RGB2BGR)
cv.imwrite("D:\Coding\img_cartoon.jpg",img_cartoon)
```
